# **Advanced visualization with seaborn**

## **First steps**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
# The following row prevent "AttributeError: 'numpy.int64' object has no attribute 'startswith'"
# https://stackoverflow.com/questions/30334428/attributeerror-numpy-int64-object-has-no-attribute-startswith
titanic.pclass = titanic.pclass.astype(str)

plt.figure(figsize = (10, 7))
sns.countplot(data = titanic, x = 'sex', hue = 'pclass')
plt.legend()
plt.show()
```

## **Categorical plots**

```python
pd.option_context('mode.use_inf_as_na', True)
plt.figure(figsize = (10, 7))
sns.stripplot(data = titanic, x = 'sex', y = 'age', jitter = 0.2, hue = 'pclass', dodge = True)
plt.show()
```

```python
plt.figure(figsize = (10, 7))
sns.swarmplot(data = titanic, x = 'sex', y = 'age', hue = 'pclass', dodge = True)
plt.show()
```

```python
plt.figure(figsize = (10, 7))
sns.violinplot(data = titanic, x = 'sex', y = 'age', hue = 'pclass', dodge = True)
plt.show()
```

```python
plt.figure(figsize = (10, 7))
sns.violinplot(data = titanic, x = 'sex', y = 'age', hue = 'pclass', dodge = True)
sns.swarmplot(data = titanic, x = 'sex', y = 'age', hue = 'pclass', dodge = True, palette = 'dark:black')
plt.show()
```

```python
plt.figure(figsize = (10, 7))
sns.violinplot(data = titanic, x = 'pclass', y = 'age', hue = 'sex', dodge = True, split = True)
plt.show()
```

```python
plt.figure(figsize = (10, 7))
sns.barplot(data = titanic, x = 'pclass', y = 'age', hue = 'sex', dodge = True)
plt.show()
```

```python
plt.figure(figsize = (10, 7))
sns.pointplot(data = titanic, x = 'pclass', y = 'age', hue = 'sex', dodge = True)
plt.show()
```

## **Jointplots/Regressions**

```python
# plt.figure(figsize = (10, 7))
sns.jointplot(data = titanic, x = 'age', y = 'fare', kind = 'reg', height = 7, ratio = 2)
plt.show()
```

```python
# plt.figure(figsize = (10, 7))
sns.lmplot(data = titanic, x = 'age', y = 'fare', height = 7, col = 'pclass')
plt.show()
```

```python
sns.lmplot(data = titanic, x = 'age', y = 'survived', height = 7, logistic = False)
plt.show()
```

```python
# plt.figure(figsize = (10, 7))
sns.lmplot(data = titanic, x = 'age', y = 'survived', height = 7, col = 'sex', logistic = True)
plt.show()
```

```python
# plt.figure(figsize = (10, 7))
sns.lmplot(data = titanic, x = 'age', y = 'survived', height = 7, col = 'pclass', logistic = True)
plt.show()
```

## **Matrixplots/Heatmaps**

```python
ct1 = pd.crosstab(titanic.sex, titanic.pclass)
```

```python
plt.figure(figsize = (10, 6))
sns.set(font_scale = 1.2)
sns.heatmap(data = ct1, annot = True, fmt = 'd', cmap = 'Blues', vmax = 150)
plt.show()
```

```python
ct2 = pd.crosstab(titanic.sex, titanic.pclass, values = titanic.survived, aggfunc = 'mean')
```

```python
ct2
```

```python
plt.figure(figsize = (10, 6))
sns.set(font_scale = 1.2)
sns.heatmap(data = ct2, annot = True, fmt = '.2f', cmap = 'Reds')
plt.show()
```

```python
corr_mat = titanic.corr(numeric_only = True)
```

```python
plt.figure(figsize = (10, 6))
sns.set(font_scale = 1.2)
sns.heatmap(data = corr_mat, annot = True, fmt = '.2f', cmap = 'Reds')
plt.show()
```
