## Indexing and slicing using **reindex()**

```python
summer_df_2 = pd.read_csv('summer.csv')
```

```python
summer_df_2.loc[[0, 5, 30000], ['Athlete', 'Medal']]
```

```python
summer_df_2.reindex(index = [0, 5, 30000, 40000], columns = ['Athlete', 'Medal'])
```

```python
summer_df_2.reindex(index = [0, 5, 30000, 40000], columns = ['Athlete', 'Age', 'Medal'])
```

```python
summer_df.reindex(index = ['PHELPS, Michael'])
```

## Advanced indexing and slicing

```python
rows = list(range(5)) + [354, 765]
rows
small_data_1 = summer_df.iloc[rows]
small_data_1
```

```python
cols = summer_df.columns[:3].to_list() + ['Gender', 'Event']
small_data_2 = summer_df.loc[:, cols]
small_data_2
```

```python
small_data_3 = summer_df_2.loc[[200, 300], ['Athlete', 'Medal']]
small_data_3
```
