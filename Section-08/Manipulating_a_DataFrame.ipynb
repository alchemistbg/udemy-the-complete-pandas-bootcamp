# **Manipulating Elements in a DataFrame**

## **Best practicies**

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

### **Change single value with _loc_**

```python
titanic.loc[1, 'age'] = 40
```

```python
titanic.head()
```

### **Change single value with _iloc_**

```python
titanic.iloc[1, 3] = 38
```

```python
titanic.head()
```

### **Change multiple values with _loc_**

```python
import numpy as np
titanic.loc[1:3, 'age'] = np.NaN
```

```python
titanic.head()
```

### **Change multiple values with _iloc_**

```python
titanic.iloc[1:4, 3] = 99
```

```python
titanic.head()
```

```python
titanic.iloc[1:4, 3] = [38, 26, 35]
```

```python
titanic.head()
```

### **Change multiple values with _boolean indexing_**

```python
babies_idx = titanic.loc[titanic.age < 1].index
```

```python
# titanic.loc[titanic.age < 1, 'age'] = 1
titanic.loc[babies_idx, 'age'] = 1
```

```python
titanic.loc[babies_idx]
```

### **Changing multiple values in a row**

```python
titanic.head()
```

```python
titanic.loc[1, 'survived':'sex'] = [0, 0, 'male']
```

```python
titanic.head()
```

### **Changing values in multiple rows and multiple columns**

```python
titanic.replace(0, 'Zero')
```

```python
titanic.head()
```

## **Chained Indexing: How you should NOT do it (Part 1)**

### **Example 1**

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
age = titanic.age
```

```python
age.head()
```

```python
age[1] = 40
```

```python
age.head()
```

```python
titanic.head()
```

```python
# Chained indexing
titanic.age[1] = 42
```

```python
titanic.age[1]
```

```python
# Non-chained indexing
titanic.loc[1, 'age'] = 38
```

```python
titanic.age[1]
```

```python
titanic.head()
```

### **Example 2**

```python
slice1 = titanic[['sex', 'age']]
```

```python
slice1.head()
```

```python
slice1.iloc[1, 1] = 43
```

```python
slice1.head()
```

```python
titanic.head()
```

### **Example 3**

```python
slice2 = titanic.loc[:, ['sex', 'age']]
```

```python
slice2.head()
```

```python
slice2.iloc[1, 1] = 42
```

```python
slice2.head()
```

```python
titanic.head()
```

## **Chained Indexing: How you should NOT do it (Part 2)**

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

### **Example 4**

```python
idx_babies = titanic[titanic.age < 1].index
```

```python
titanic[titanic.age < 1]['age'] = 1
```

```python
titanic.loc[idx_babies, :]
```

### **Example 5**

```python
titanic['age'][titanic.age < 1] = 1
```

```python
titanic.loc[idx_babies, :]
```

### **Example 6**

```python
titanic[['sex', 'age']][titanic.age == 1]['age'] = 0
```

```python
titanic.loc[idx_babies]
```

## **View vs. Copy**

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
age = titanic.age
```

```python
age._is_view
```

```python
age._is_copy is None
```

```python
age[1] = 42
```

```python
age.head()
```

```python
titanic.head()
```

```python
babies = titanic[titanic.age < 1]
```

```python
babies
```

```python
babies._is_view
```

```python
babies._is_copy is None
```

```python
babies._is_copy()
```

```python
babies.age = 1
```

```python
babies
```

```python
titanic.loc[babies.index]
```

### **Manipulating a Slice**

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
age = titanic.age.copy()
```

```python
age._is_view
```

```python
age._is_copy is None
```

```python
age.head()
```

```python
age[1] = 42
```

```python
age.head()
```

```python
titanic.head()
```

```python
babies = titanic.loc[titanic.age < 1, ['age', 'sex']].copy()
```

```python
babies['age'] = 1
```

```python
babies
```

```python
titanic.loc[babies.index]
```
