## **_JOIN_ oprations**

#### **Outer JOIN**

```python
men_2004_3 = pd.read_csv('men2004.csv')
```

```python
men_2004_3.head()
```

```python
men_2004_3.shape
```

```python
men_2008_3 = pd.read_csv("men2008.csv")
```

```python
men_2008_3.head()
```

```python
men_2008_3.shape
```

```python
men_2004_3.merge(right = men_2008_3, how = 'outer', on = 'Athlete')
```

```python
men_2004_3.merge(right = men_2008_3, how = 'outer', on = 'Athlete', suffixes = (' - 2004', ' - 2008'))
```

```python
men_2004_3.merge(right = men_2008_3, how = 'outer', on = 'Athlete', suffixes = (' - 2004', ' - 2008'), indicator = 'Source')
```

```python
men_0408_03 = men_2004_3.merge(
    right = men_2008_3,
    how = 'outer',
    on = 'Athlete',
    suffixes = (' - 2004', ' - 2008'),
    indicator = 'Source'
)
```

```python
men_0408_03
```

```python
men_0408_03.Source.value_counts()
```
