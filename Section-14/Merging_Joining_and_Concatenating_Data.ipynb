## **JOINing on different columns**

```python
men_2004.head()
```

```python
men_2008.head()
```

```python
men_2004_4 = men_2004.copy()
```

```python
men_2008_4 = men_2008.copy()
```

```python
men_2008_4.columns = ['Athlete', 'Medals']
```

```python
men_0408_2 = men_2004_4.merge(
    right = men_2008_4,
    how = 'outer',
    left_on = 'Name', right_on = 'Athlete',
    suffixes = (' - 2004', ' - 2008'),
    indicator = True
)
```

```python
men_0408_2.Name.fillna(men_0408_2.Athlete, inplace = True)
```

```python
men_0408_2
```

```python
men_0408_2.drop(columns = 'Athlete', inplace = True)
```

```python
men_0408_2
```

```python
men_0408_2.rename(columns = {'Medals - 2004': 'Medals_2004', 'Medals - 2008': 'Medals_2008'}, inplace = True)
```

```python
men_0408_2.Medals_2004.fillna(0, inplace = True)
```

```python
men_0408_2.Medals_2008.fillna(0, inplace = True)
```

```python
men_0408_2
```

```python
men_2008_4.set_index('Athlete', inplace = True)
```

```python
men_0408_3 = men_2004_4.merge(
    right = men_2008_4,
    how = 'outer',
    left_on = 'Name', right_index = True,
    suffixes = ('_2004', '_2008'),
    indicator = True
)
```

```python
men_0408_3
```
