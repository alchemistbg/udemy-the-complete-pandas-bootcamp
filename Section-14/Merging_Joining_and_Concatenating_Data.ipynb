# **Merging, Joining and Concatenating Data**

## **Adding/Concatenating rows**

```python
import pandas as pd
```

```python
men_2004 = pd.read_csv('men2004.csv')
```

```python
men_2004
```

```python
men_2008 = pd.read_csv('men2008.csv')
```

```python
men_2008
```

**dataframe.append()** function is deprecated

```python
pd.concat([men_2004, men_2008])
```

```python
pd.concat([men_2004, men_2008], ignore_index = True)
```

```python
pd.concat([men_2004, men_2008], ignore_index = True, axis = 0)
```

```python
men_0408 = pd.concat([men_2004, men_2008], keys = [2004, 2008], names = ['Year', 'Idx'])
```
