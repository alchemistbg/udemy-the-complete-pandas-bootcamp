## **Measuring stock performance using MEAN return and STD of Return**

```python
aapl_close = close.AAPL.copy().to_frame()
```

```python
aapl_close.pct_change().dropna()
```

```python
aapl_return = aapl_close.pct_change().dropna().mul(100)
```

```python
aapl_return.info()
```

```python
aapl_return.head()
```

```python
aapl_return.plot(kind = 'hist', bins = 100)
plt.show()
```

```python
return_mean = aapl_return.mean()
```

```python
return_std = aapl_return.std()
```

```python
return_var = aapl_return.var()
```

```python
print(f'Mean: {return_mean.values[0]:.4f}\nStd dev: {return_std.values[0]:.4f}\nVar: {return_var.values[0]:.4f}')
```
