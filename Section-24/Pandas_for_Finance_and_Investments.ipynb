## **Financial time series - Return and Risk**

```python
import numpy as np
```

```python
close.head()
```

```python
close.pct_change().dropna().mul(100)
```

```python
return_all = close.pct_change().dropna().mul(100)
```

```python
return_all
```

```python
return_all.describe()
```

```python
summary_all = return_all.describe().T.loc[:, ['mean', 'std']]
```

```python
summary_all
```

```python
summary_all['mean'] = summary_all['mean'].mul(252)
```

```python
summary_all['std'] = summary_all['std'].mul(np.sqrt(252))
```

```python
summary_all
```

```python
summary_all.plot(kind = 'scatter', x = 'std', y = 'mean',)
plt.xlabel('Annual risk', weight = 'bold')
plt.ylabel('Annual return', weight = 'bold')
plt.title('Risk/Return', fontsize = 14, weight = 'bold')
for idx in summary_all.index:
    plt.annotate(text = idx, xy = (summary_all.loc[idx, 'std'] + 0.3, summary_all.loc[idx, 'mean'] - 0.25))
plt.grid()
plt.show()
```
