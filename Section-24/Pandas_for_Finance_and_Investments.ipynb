## **Import and export stock price data from Yahoo**

```python
# %%HTML
# <style type="text/css">
#     table.dataframe td, table.dataframe th {
#         border-style: solid;
#     }
# </style>
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
ticker = ['AAPL', 'BA', 'KO', 'IBM', 'DIS', 'MSFT']
```

```python
stocks = yf.download(ticker, start = '2015-01-01', end = '2024-04-30')
```

```python
stocks
```

```python
stocks.head()
```

```python
stocks.tail()
```

```python
stocks.loc[:, (['High', 'Low'], 'AAPL')].plot()
plt.show()
```

```python
stocks.to_csv('stocks_2015-2024.csv')
```

```python
stocks_from_csv = pd.read_csv('stocks_2015-2024.csv', header = [0, 1], index_col = [0], parse_dates = [0])
```
