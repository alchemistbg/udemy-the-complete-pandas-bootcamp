## **Customization of the plots**

```python
plt.style.available
```

```python
plt.style.use('fast')
```

```python
titanic.age.plot(figsize = (10, 5), c = 'red', xlim = (0, 900), ylim = (0, 80))
plt.title('Titanic - Ages', fontsize = 15)
plt.legend(loc = 'best')
plt.xlabel('Passengers', fontsize = 13)
# plt.xticks(ticks = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900])
plt.ylabel('Age', fontsize = 13)
plt.grid(visible = True)
plt.show()
```

## **Histograms**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8')
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
titanic.describe()
```

```python
titanic.age.value_counts()
```

```python
titanic.age.value_counts(bins = 10)
```

```python
titanic.age.plot(kind = 'line', figsize = (10, 6))
plt.show()
```

```python
titanic.age.plot(kind = 'hist', figsize = (10, 6), bins = 80, rwidth = 0.75)
plt.show()
```

```python
titanic.age.hist(bins = 80, figsize = (10, 6), rwidth = 0.75)
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
plt.hist(titanic.age, bins = 80, rwidth = 0.75)
plt.show()
```

## **Barcharts and Piecharts**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8')
```

```python
summer_2012 = pd.read_csv('summer_2012.csv', index_col = 'Country')
```

```python
summer_2012.head()
```

```python
summer_2012.Medal.plot(kind = 'bar', figsize = (10, 6))
plt.show()
```

```python
summer_2012.Medal.plot(kind = 'barh', figsize = (10, 6))
plt.show()
```

```python
summer_2012.Medal.plot(kind = 'pie', figsize = (10, 6), legend = True, labels = None)
plt.legend(labels = summer_2012.index, loc = 'upper left', bbox_to_anchor=(1.0, 1))
plt.show()
```

## **Scatterplot**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('seaborn-v0_8')
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
titanic.plot(kind = 'scatter', figsize = (10, 6), x = 'age', y = 'fare')
plt.show()
```

```python
titanic.plot(kind = 'scatter', figsize = (10, 6), x = 'age', y = 'fare', c = 'pclass', colormap = 'viridis', marker = 'D')
plt.show()
```
